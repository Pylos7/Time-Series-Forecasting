```python
#libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import pmdarima as pm
```

```python
#load the data
#YYYY-MM-DD
df = pd.read_csv('../nyc_data.csv', index_col = 0, parse_dates = True)
future_df = pd.read_csv('../future.csv', index_col = 0, parse_dates = True)
future_df.head()
```

```python
#Rename variable
df = df.rename(columns = {'Demand': 'y'})
df.head(1)
```

```python
# Extract regressors
train_X = df.iloc[:,1:]
# All values in axis 0 and values from the second column and on from axis 1
future_X = future_df.iloc[:,1:]
future_X.head()
```

# SARIMAX Model

```python
# Get the best parameters
parameters = pd.read_csv("../Forecasting Product/best_params_sarimax.csv", index_col = 0)
parameters
```

```python
# Store the individual parameters
p = parameters.loc["p"][0]
d = parameters.loc["d"][0]
q = parameters.loc["q"][0]
P = parameters.loc["P"][0]
D = parameters.loc["D"][0]
Q = parameters.loc["Q"][0]
```

```python
# Model
# hourly: 24, daily: 7, weekly:52, monthly:12, quarterly:4
model = pm.ARIMA(order = (p, d, q),
                 seasonal_order = (P, D, Q, 7),
                 suppress_warnings = True,
                 enforce_stationarity  = False)
model.fit(df.y)
```

# Forecasting

```python
# Predictions
predictions_sarimax = model.predict(n_periods = len(future_df),
              X = future_X).rename("sarimax")
```

```python
# Visualization
df['y']['2020-01-01':].plot(figsize = (15, 6), legend = True);
predictions_sarimax.plot(legend = True);
```

```python
# Exporting
predictions_sarimax.to_csv("Ensemble/predictions_sarimax.csv")
```
